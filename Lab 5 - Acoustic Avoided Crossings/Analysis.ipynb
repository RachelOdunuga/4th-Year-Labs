```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
def find_resonances(df, channel="A_ch0", prominence=0.001):
    y = df[channel].values
    peaks, props = find_peaks(y, prominence=prominence)
    return df["f_kHz"].iloc[peaks].values, y[peaks]
```
